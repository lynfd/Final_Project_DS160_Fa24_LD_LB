# How does proficiency in reading at the end of primary education and proficiency in math at the end of primary education  affect unemployment rates?

```python
# importing necessary data for linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
X = data[['Primary_End_Proficiency_Reading', 'Primary_End_Proficiency_Math']]
Y = data['Unemployment_Rate']
```

```python
X.sample()
```

```python
# Split data into training and testing sets
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

# Initialize and train the model
mlr_model=LinearRegression()
mlr_model.fit(x_train,y_train)

# Make predictions on the test set
y_pred_mlr=mlr_model.predict(x_test)
```

```python
mse=mean_squared_error(y_test, y_pred_mlr)
```

```python
# Evaluate the model
mse_mlr=mean_squared_error(y_test,y_pred_mlr)
rmse_mlr=np.sqrt(mse)
r2_mlr=r2_score(y_test,y_pred_mlr)

# Output evaluation metrics
print("Multiple Linear Regression Evaluation Metrics:")
mse_mlr,rmse_mlr,r2_mlr
```

### Given the unusually high mean squared error, we have found that there is not a linear relationship between the independent variables "Primary_End_Proficient_Reading", and "Primary_End_Proficiency_Math", and the dependant variable "Unemployement_Rate".
### Thus, there is not a linear relationship between reading and math proficency and the unemployment rate.

# Can we classify a person as literate if their proficiency in reading is above a certain number at the end of primary or lower secondary education? What is that number?â€‹

```python
# Install the required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn import utils
warnings.filterwarnings('ignore')
#Load the data set
data=pd.read_csv('Global_Education.csv', encoding='latin-1')
#visualize the top five rows in the data using the head function
data.head()
```

```python
# calculate the average literacy rate and create a new colum
data['Youth_15_24_Literacy_Rate_Avg'] = (data['Youth_15_24_Literacy_Rate_Male'] + data['Youth_15_24_Literacy_Rate_Female']) / 2
```

```python
data['Developed_Country']=np.where(data['Youth_15_24_Literacy_Rate_Avg']<95, 'Not Developed', 'Developed')
data.head()
```

```python
df=data[['Developed_Country','Continent','Birth_Rate','Gross_Primary_Education_Enrollment','Gross_Tertiary_Education_Enrollment']]
df.head()
```

```python
df.isnull().sum()
```

```python
df['Continent'].unique()
```

```python
lab=preprocessing.LabelEncoder()
x=df.drop('Developed_Country',axis=1)
y=df["Developed_Country"]
```

```python
df.info()
```

```python
df['Birth_Rate']=df['Birth_Rate'].astype('int64')
df['Gross_Primary_Education_Enrollment']=df['Gross_Primary_Education_Enrollment'].astype('int64')
df['Gross_Tertiary_Education_Enrollment']=df['Gross_Tertiary_Education_Enrollment'].astype('int64')
```

```python
x=pd.get_dummies(x, drop_first=True)
x.head()
```

```python
# Splitting the data between train and test set
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# Train the logistic Regression Model
LR=LogisticRegression(max_iter=1000)
LR.fit(x_train,y_train)

# Make prediction
y_pred=LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Confusion matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
```

```python
print(classification_report(y_test,y_pred))
```
